```python
# code to check whether a string is palindrome or not
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

word = input("Enter a string: ")
if is_palindrome(word):
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# code to check whether a string is palindrome or not
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])

text = input("Enter a string: ")
if is_palindrome(text):
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# Generate fibonacci sequence
# using recursion
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)

n = int(input("Enter number of terms: "))
for i in range(n):
    print(fib(i), end=" ")
```

```python
# generate fibonacci sequence
# using dynamic programming
fib_cache = {}

def fib(n):
    if n in fib_cache:
        return fib_cache[n]
    if n <= 1:
        value = n
    else:
        value = fib(n-1) + fib(n-2)
    fib_cache[n] = value
    return value

n = int(input("Enter number of terms: "))
for i in range(n):
    print(fib(i), end=" ")
```

```python
# adding two numbers without using '+' operator
def add(a, b):
    while b != 0:
        carry = a & b        # carry bits
        a = a ^ b            # add without carry
        b = carry << 1       # shift carry to add in next iteration
    return a

# Example
print(add(10, 20))   # Output: 30
```

```python
# Factorial Using Recursion (with Negative Number Handling)
def factorial(n):
    if n < 0:
        return "Factorial does not exist for negative numbers"
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# Example
print(factorial(5))
print(factorial(-3))
```

```python
def find_largest(lst):
    if not lst:
        return "List is empty"

    largest = lst[0]   # Assume first element is largest

    for num in lst:
        if num > largest:
            largest = num

    return largest

# Example
print(find_largest([10, 45, 23, 89, 2, 77]))  # Output: 89
```

```python
# To calculate the frequency of each element in a list
def frequency_count(lst):
    freq = {}

    for item in lst:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    return freq

# Example
print(frequency_count([1, 2, 2, 3, 3, 3, 4]))
# Output: {1: 1, 2: 2, 3: 3, 4: 1}
```

```python
# To find the second largest element in a list
def second_largest(lst):
    if len(lst) < 2:
        return "List must contain at least two elements"

    largest = second_largest = float('-inf')

    for num in lst:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest

# Example
print(second_largest([10, 45, 23, 89, 2, 77]))  # Output: 77
```

```python
# to remove duplicates from a list while preserving the original order
def remove_duplicates(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

# Example
print(remove_duplicates([1, 2, 2, 3, 4, 4, 5]))
# Output: [1, 2, 3, 4, 5]
```

```python
# to swap three numbers with each other
def swap_three(a, b, c):
    temp = a
    a = b
    b = c
    c = temp
    return a, b, c

# Example
print(swap_three(10, 20, 30))
# Output: (20, 30, 10)
```
